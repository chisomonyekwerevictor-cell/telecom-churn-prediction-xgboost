```python
 df = df.drop(columns=['CustomerID', 'Count', 'Country', 'State', 'City', 'Zip Code','Lat Long', 'Latitude', 'Longitude', 'Churn Score', 'Churn Reason', 'CLTV', 'Churn Value'], errors="ignore")
```

```python
# List all columns that should be numeric
numeric_columns = ["Total Charges", "Monthly Charges", "Tenure Months"]

# Convert them all to numeric
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    # Fill any NaN that might appear
    df[col] = df[col].fillna(df[col].median())

# Check they're now numeric
print(df[numeric_columns].dtypes)
```

```python
# Check what Contract looks like
print(df["Contract"].unique())
```

```python
# Encode Contract as ordinal
df["Contract"] = (df["Contract"]
                  .astype(str)
                  .str.strip()
                  .str.lower()
                  .map({"month-to-month": 0, "one year": 1, "two year": 2}))

print("Contract encoded:", df["Contract"].unique())
print("Contract NaNs:", df["Contract"].isnull().sum())
```

```python
target_col = ['Churn Label']


###  Get all categorical columns (excluding the target)
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
if target_col in categorical_cols:
    categorical_cols.remove(target_col)

# Separate binary from nominal columns
binary_cols = []
nominal_cols = []

for col in categorical_cols:
    if df[col].nunique() == 2:
        binary_cols.append(col)
    else:
        nominal_cols.append(col)

print("Binary columns:", binary_cols)
print("Nominal columns:", nominal_cols)
```

```python
#### Encoding binary columns (manual mapping)###

binary_map = {"Yes": 1,"No": 0,"Male": 1,"Female": 0}

for col in binary_cols:
    df[col] = df[col].map(binary_map)
```

```python
print("Nominal columns:", nominal_cols)
```

```python
print(set(nominal_cols) - set(df.columns))
```

```python
nominal_cols = [col for col in nominal_cols if col in df.columns]
```

```python
# Step 5: One-hot encode nominal (only if any exist)
if nominal_cols:
    df = pd.get_dummies(df, columns=nominal_cols, drop_first=True)
else:
    print("No nominal columns to encode")

print("\nEncoding complete!")
print("Final shape:", df.shape)
```

```python
### spliting the data
from sklearn.model_selection import train_test_split
X = df.drop(columns=['Churn Label'])
y = df['Churn Label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
def sanity_check(X, y):
    print("SANITY CHECK START \n")
    
    # 1. Shape check
    print("Shapes:")
    print(f"X shape: {X.shape}")
    print(f"y shape: {y.shape}\n")
    
    assert X.shape[0] == y.shape[0], "❌ Row mismatch between X and y"
    
    # 2. Target check
    print("Target distribution:")
    print(y.value_counts())
    print()
    
    # 3. Missing values
    na_count = X.isna().sum().sum()
    print(f"Total NaNs in X: {na_count}")
    assert na_count == 0, "❌ NaNs still present in X"
    
    # 4. Data types
    non_numeric = X.select_dtypes(include=["object"]).columns.tolist()
    print(f"Non-numeric columns in X: {non_numeric}")
    assert len(non_numeric) == 0, "❌ Non-numeric columns found"
    
  
    # 6. Index alignment
    assert (X.index == y.index).all(), "❌ X and y indices are misaligned"
    
    print("\n✅ SANITY CHECK PASSED — DATA IS MODEL READY")
    print("SANITY CHECK END")
```

```python
sanity_check(X, y)
```

```python
#### using  a baseline model###
from sklearn.linear_model import LogisticRegression


lr = LogisticRegression(random_state=42, max_iter=3000, class_weight ="balanced")
lr.fit(X_train, y_train) 
y_pred = lr.predict(X_test)
print("Logistic Regression Results")
print(classification_report(y_test, y_pred))
```

## Threshold Selection Rationale ###

#### Multiple decision thresholds were evaluated to balance recall and precision ###
### for the minority class. Since the cost of false negatives is high, recall was ###
### prioritized over raw accuracy. ###

#### A threshold of *0.35* was selected as it provides strong recall (0.88) for ###
### the minority class while maintaining reasonable precision and overall model ###
#### stability. ####
